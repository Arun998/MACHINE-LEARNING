```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
df.head()
```

```python
df['target']=cancer.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df['cancer_class']=df.target.apply(lambda x: cancer.target_names[x])
df.head()
```

```python
df0=df[df.target==0].head(50)
df1=df[df.target==1].head(50)
print(df0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("mean radius")
plt.ylabel("mean texture")
plt.scatter(df0['mean radius'],df0['mean texture'],color='red',marker='+')
plt.scatter(df1['mean radius'],df1['mean texture'],color='blue',marker='*')
```

```python
X=cancer.data
Y=cancer.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
mymodel=SVC(C=2,kernel='linear')
```

```python
mymodel.fit(X_train,Y_train)
```

```python
mymodel.score(X_test,Y_test)
```

```python
pred=mymodel.predict(X_test)
print(pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,pred)
cm
```

```python
model_kernal=SVC(C=3,kernel='linear')
model_kernal.fit(X_train,Y_train)
model_kernal.score(X_test,Y_test)
```
